```python
drop_cols = [
    'animal_id','name','count','date_of_birth',
    'monthyear','dob_monthyear','sex_age_outcome',
    'outcome_age_(days)','outcome_age_(years)',
    'Cat/Kitten (outcome)'
]

acc_cat.drop(drop_cols, axis=1, inplace=True)
```

```python
# 1️⃣ Drop missing target
acc_cat = acc_cat.dropna(subset=['target'])

# 2️⃣ Fill categorical missing values
acc_cat['outcome_subtype'] = acc_cat['outcome_subtype'].fillna('None')
acc_cat['color'] = acc_cat['color'].fillna('Unknown')
acc_cat['breed2'] = acc_cat['breed2'].fillna('No Mix')
acc_cat['coat_pattern'] = acc_cat['coat_pattern'].fillna('Unknown')
acc_cat['color2'] = acc_cat['color2'].fillna('No Second Color')

# 3️⃣ Final check
acc_cat.isnull().sum()
```

```python
X = acc_cat.drop("target", axis=1)
y = acc_cat["target"]
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(handle_unknown="ignore", drop="first"), cat_cols)
    ],
    remainder="passthrough"
)

X_encoded = ct.fit_transform(X)
```

```python
le = LabelEncoder()

# target encode
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
# Categorical columns
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()
```

```python

# pipeline
pipeline = Pipeline([
    ("encoder", ColumnTransformer(
        transformers=[
            ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
        ],
        remainder="passthrough"
    )),
    ("model", RandomForestClassifier(random_state=42))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
print("Accuracy:", pipeline.score(X_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(pipeline, "shelter_outcome_pipeline.pkl")
```
